```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('all_spectra_row_v1.csv', index_col=0)
```

```python
df.groupby('label').count()
```

```python
## Creating cases with each label with 30 instances in each

## Has alreay 30 instances
cases_11 = df[df['label'] == 1.1]

# 30 instances will remain in the training set, rest will go to the test set
cases_111 = df[df['label'] == 1.11][:30]
cases_111_test = df[df['label'] == 1.11][30:]

# 30 instances will remain in the training set, rest will go to the test set
cases_15 = df[df['label'] == 1.5][:30]
cases_15_test = df[df['label'] == 1.5][30:]

# 30 instances will remain in the training set, rest will go to the test set
cases_16 = df[df['label'] == 1.6][:30]
cases_16_test = df[df['label'] == 1.6][30:]

# To get 30 instances the missing amount is produced by duplicating the existing instances
cases_19 = df[df['label'] == 1.9]
cases_19 = pd.concat([cases_19, cases_19[:30 - len(cases_19.index)]])

# 30 instances will remain in the training set, rest will go to the test set
cases_21 = df[df['label'] == 2.1][:30]
cases_21_test = df[df['label'] == 2.1][30:]

# To get 30 instances the missing amount is produced by duplicating the existing instances
cases_23 = df[df['label'] == 2.3]
cases_23 = pd.concat([cases_23, cases_23[:30 - len(cases_23.index)]])

# 30 instances will remain in the training set, rest will go to the test set
cases_31 = df[df['label'] == 3.1][:30]
cases_31_test = df[df['label'] == 3.1][30:]

# To get 30 instances the missing amount is produced by duplicating the existing instances
cases_311 = df[df['label'] == 3.11]
cases_311 = pd.concat([cases_311, cases_311[:30 - len(cases_311.index)]])

# To get 30 instances the missing amount is produced by duplicating the existing instances
cases_312 = df[df['label'] == 3.12]
cases_312 = pd.concat([cases_312, cases_312[:30 - len(cases_312.index)]])

# To get 30 instances the missing amount is produced by duplicating the existing instances
cases_32 = df[df['label'] == 3.2]
cases_32 = pd.concat([cases_32, cases_32[:30 - len(cases_32.index)]])

# 30 instances will remain in the training set, rest will go to the test set
cases_35 = df[df['label'] == 3.5][:30]
cases_35_test = df[df['label'] == 3.5][30:]


df_train = pd.concat([cases_11, cases_111, cases_15, cases_16, cases_19, cases_21, cases_23, cases_31, cases_311, cases_312, cases_32, cases_35])
df_test = pd.concat([cases_111_test, cases_15_test, cases_16_test, cases_21_test, cases_31_test, cases_35_test])

## Writing to csv
df_train.to_csv('spectra_balanced.csv')
df_test.to_csv('spectra_test.csv')
```
